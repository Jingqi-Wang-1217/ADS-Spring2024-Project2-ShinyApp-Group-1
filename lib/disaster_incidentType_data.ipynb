```python
import pandas as pd

df=pd.read_csv("DisasterDeclarationsSummaries.csv")
selected_df = df[['disasterNumber','state','fyDeclared','incidentType','hmProgramDeclared']].drop_duplicates()

labels = {
    'Northeast': ['CT', 'ME', 'MA', 'NH', 'RI', 'VT', 'NJ', 'NY', 'PA', "DE"],
    'Southeast': ['FL', 'GA', 'MD', 'NC', 'SC', 'VA', 'DC', 'WV', "TN", "AL", "KY", "MS", 'VI'],
    'Midwest': ['IL', 'IN', 'IA', 'KS', 'MI', 'MN', 'MO', 'NE', 'ND', 'OH', 'SD', 'WI'],
    'Southwest': ['AR', 'AZ', 'LA', 'NM', 'OK', 'TX'],
    'West': ['AK', 'CA', 'CO', 'HI', 'ID', 'MT', 'NV', 'OR', 'UT', 'WA', 'WY'],
    'Territories': ['AS', 'GU', 'MP', 'PR', "FM"]
}

# Add "Region" column to the df
selected_df['Region'] = selected_df['state'].map(
    {state: region for region, states in labels.items() for state in states})


selected_df.to_csv("disaster_incidentType.csv")
```
